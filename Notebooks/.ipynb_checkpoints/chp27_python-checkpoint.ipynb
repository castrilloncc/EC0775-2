### Example: Married Women's Annual Labor Supply

The file `mroz.dta` includes data on hours worked for 753 married women, 428 of whom worked for a wage outside the home during the year; 325 of the women worked zero hours. For the women who worked positive hours, the range is fairly broad, extending from 12 to 4,950. Thus, annual hours worked is a good candidate for a Tobit model.

````python
%%stata
use "http://fmwww.bc.edu/ec-p/data/wooldridge/mroz.dta", clear
list hours nwifeinc educ exper age kidslt6 kidsge6 in 426/431, clean 
````

````python
%%stata
set scheme s1mono
histogram hours, start(0) width(1) xtitle(Annual Hours Worked)
````

````python
%%stata
list hours nwifeinc educ exper age kidslt6 kidsge6 in 423/428 if hours>0, clean 
````

````python
%%stata
set scheme s1mono
histogram hours if hours>0, xtitle(Yearly Hours Worked)
````

## Censored/Truncated Regression

````python
from pystata import stata
stata.run('quietly: collect _r_b _r_se _r_p, tag(model[Linear (OLS)]): reg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, robust')
stata.run('quietly: collect _r_b _r_se _r_p, tag(model[Tobit (MLE)]): tobit hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)')
stata.run('quietly: collect _r_b _r_se _r_p, tag(model[Truncated (MLE)]): truncreg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)')
stata.run('collect style cell, nformat(%5.3f)')
stata.run('collect style cell border_block, border(right, pattern(nil))')
stata.run('collect style cell result[_r_se], sformat("(%s)")')
stata.run('collect style header result, level(hide)')
stata.run('collect style column, extraspace(1)')
stata.run('collect style row stack, spacer delimiter(" x ")')
stata.run('collect recode colname c.exper#c.exper = exper^2')
stata.run('collect recode colname _cons = Constant')
stata.run('collect layout (colname#result) (model)')
````

## Censored/Truncated Expectations

````python
stata.run('quietly reg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, robust')
stata.run('collect _r_b _r_se _r_p, tag(modelo[Linear (OLS)]): margins, atmeans')
stata.run('quietly tobit hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, vce(robust) ll(0)')
stata.run('quietly: collect _r_b _r_se _r_p, tag(modelo[Tobit (MLE)]): margins, atmeans')
stata.run('quietly truncreg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, vce(robust) ll(0)')
stata.run('quietly: collect _r_b _r_se _r_p, tag(modelo[Truncated (MLE)]): margins, atmeans')
stata.run('collect style cell, nformat(%5.3f)')
stata.run('collect style cell border_block, border(right, pattern(nil))')
stata.run('collect style cell result[_r_se], sformat("(%s)")')
stata.run('collect style header result, level(hide)')
stata.run('collect style column, extraspace(1)')
stata.run('collect style row stack, spacer delimiter(" x ")')
stata.run('collect recode colname _cons = E[Y|X]')
stata.run('collect layout (colname#result) (modelo)')
````

````python
%%stata
quietly reg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, robust
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(a_modelo[Linear (OLS)]):
        margins, at(kidslt6=1 kidsge6=0  (mean) nwifeinc educ exper age);
#delimit cr
quietly tobit hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(a_modelo[Tobit (MLE)]):
        margins, at(kidslt6=1 kidsge6=0  (mean) nwifeinc educ exper age);
#delimit cr
quietly truncreg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(a_modelo[Truncated (MLE)]):
        margins, at(kidslt6=1 kidsge6=0  (mean) nwifeinc educ exper age);
#delimit cr
````

👉🏼 The Linear model tries to estimate the expected number of _desired_ hours of work.

👉🏼 The Truncated model estimates the expected number of _actual_ hours of work for workers.

👉🏼 The Tobit model estimates the expected number of _actual_ hours of work for workers and nonworkers.

For a 30 years old woman who is a college graduate, has 5 years of experience, has 1 kid less than 6 years old and whose partner earns roughly US$20K a year.

````python
stata.run('collect style cell, nformat(%5.3f)')
stata.run('collect style cell border_block, border(right, pattern(nil))')
stata.run('collect style cell result[_r_se], sformat("(%s)")')
stata.run('collect style header result, level(hide)')
stata.run('collect style column, extraspace(1)')
stata.run('collect style row stack, spacer delimiter(" x ")')
stata.run('collect recode colname _cons = E[Y|X]')
stata.run('collect layout (colname#result) (a_modelo)')
````

````python
%%stata
quietly reg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, robust
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(b_modelo[Linear (OLS)]):
        margins, at(kidslt6=0 kidsge6=0  (mean) nwifeinc educ exper age);
#delimit cr
quietly tobit hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(b_modelo[Tobit (MLE)]):
        margins, at(kidslt6=0 kidsge6=0  (mean) nwifeinc educ exper age);
#delimit cr
quietly truncreg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(b_modelo[Truncated (MLE)]):
        margins, at(kidslt6=0 kidsge6=0  (mean) nwifeinc educ exper age);
#delimit cr
````

For a 30 years old woman who is a college graduate, has 5 years of experience, has no kid less than 6 years old and whose partner earns roughly US$20K a year.

````python
stata.run('collect style cell, nformat(%5.3f)')
stata.run('collect style cell border_block, border(right, pattern(nil))')
stata.run('collect style cell result[_r_se], sformat("(%s)")')
stata.run('collect style header result, level(hide)')
stata.run('collect style column, extraspace(1)')
stata.run('collect style row stack, spacer delimiter(" x ")')
stata.run('collect recode colname _cons = E[Y|X]')
stata.run('collect layout (colname#result) (b_modelo)')
````

⚠ It seems there is an issue regarding the notation and command options Stata uses to calculate marginal effects after a censored and truncated regression. See [this paper](https://mpra.ub.uni-muenchen.de/33252/) and [this post](https://www.statalist.org/forums/forum/general-stata-discussion/general/1531196-tobit-marginal-effects). The following code implement the equations discussed in the lecture notes.

````python
%%stata
quietly tobit hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(c_modelo[Tobit (MLE)]):
                                  margins, dydx(*)  predict(ystar(0,.));
#delimit cr
quietly truncreg hours nwifeinc educ exper c.exper#c.exper age kidslt6 kidsge6, ll(0)
#delimit ;
quietly: collect _r_b _r_se _r_p, tag(c_modelo[Truncated (MLE)]):
                                  margins, dydx(*) predict(e(0,.));
#delimit cr
````

# Sample Selection

### Example: Wage Offer Equation for Married Women

We apply the sample selection correction to the data on married women in `mroz.dta`. Recall that of the 753 women in the sample, 428 worked for a wage during the year. The wage offer equation is standard, with $log⁡(wage)$ as the dependent variable and $educ$, $exper$, and $exper^2$ as the explanatory variables. The fact that $nwifeinc$, $age$, $kidslt6$, and $kidsge6$ are excluded from the wage offer equation is an _assumption_, i.e., we assume that, given the productivity factors, these factors have no effect on the wage offer, **but** they are believed to affect the decision to work for wages.

````python
%%stata
use "http://fmwww.bc.edu/ec-p/data/wooldridge/mroz.dta", clear
list lwage educ exper nwifeinc age kidslt6 kidsge6 in 426/431, clean 
````

## Heckman's Model (Maximum Likelihood Estimator)

````python
%%stata
heckman lwage educ exper c.exper#c.exper, select(nwifeinc age kidslt6 kidsge6) robust mle
````

## Heckman's Model (Two-Step)

````python
%%stata
heckman lwage educ exper c.exper#c.exper, select(nwifeinc age kidslt6 kidsge6) twostep
````
